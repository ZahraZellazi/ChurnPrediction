```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
```

```python
#1-chargement des donn√©es
data_path = "churn-bigml-80.csv" 
data = pd.read_csv(data_path)
```

```python
data.head()
```

```python
#2.Exploration des donn√©es : EDA
print(data.shape)
print(data.columns)
print(data.dtypes)
```

```python
#D√©tection des valeurs manquantes
print('valeurs manquantes =\n',data.isnull().sum())
```

```python
#categorical cols
categorical_cols = data.columns[data.dtypes==object].tolist()
print('categorical cols = \n',categorical_cols)
print('categorical cols = \n',len(categorical_cols))
print('categorical cols(modalities) = \n',data[categorical_cols].nunique())
```

```python
#numerical cols
numerical_cols = data.columns[data.dtypes!=object].tolist()
print('numerical cols = \n',numerical_cols)
print('numerical cols = \n',len(numerical_cols))
```

```python
#description des variables quantitatives
print(data.describe())
```

```python

# Set the style for the plots
sns.set(style="whitegrid")

# Improved plots with x-axis data swapped to the y-axis and vice versa
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    
    # Create a countplot with swapped axes
    sns.countplot(x=data[col], orient="v", order=data[col].value_counts().index, palette="Set2")
    
    # Adding title and labels with better formatting
    plt.title(f'Distribution of {col}', fontsize=10, fontweight='bold', color='darkblue')
    plt.xlabel('Count', fontsize=12, fontweight='bold', color='darkgreen')
    plt.ylabel(col, fontsize=12, fontweight='bold', color='darkgreen')
    
    # Rotate tick labels on the x-axis if needed
    plt.xticks(rotation=45, fontsize=10, ha="right")
    
    # Add annotations for each bar
    for p in plt.gca().patches:
        plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                           ha='center', va='bottom', fontsize=6, color='black', fontweight='bold', xytext=(0, 10), textcoords='offset points')
    
    # Show the plot
    plt.tight_layout()  # Adjust layout to prevent clipping of labels
    plt.show()
```

```python

# Set a consistent style for all plots
sns.set(style="whitegrid")

# Enhanced plots for numerical features
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    
    # Create histogram with KDE and enhanced aesthetics
    sns.histplot(data[col], kde=True, color="teal", edgecolor="black", linewidth=1.2, bins=30)
    
    # Add title and axis labels with improved formatting
    plt.title(f'Distribution of {col}', fontsize=16, fontweight='bold', color='darkblue')
    plt.xlabel(col, fontsize=12, fontweight='bold', color='darkgreen')
    plt.ylabel('Frequency', fontsize=12, fontweight='bold', color='darkgreen')
    
    # Customize grid and background
    plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
    
    # Add annotations for peak values (max density point in KDE)
    if data[col].notnull().any():  # Ensure the column is not empty
        kde_ax = sns.kdeplot(data[col], color="blue", linewidth=2)
        max_density = kde_ax.lines[0].get_data()[1].max()  # Find max y-value (density)
        max_x = kde_ax.lines[0].get_data()[0][kde_ax.lines[0].get_data()[1].argmax()]  # Corresponding x-value
        plt.annotate(f"Peak: {max_x:.2f}", xy=(max_x, max_density), xytext=(max_x + 0.5, max_density + 0.02),
                     arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color="black", fontweight="bold")
    
    # Adjust layout to avoid clipping
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a copy of the dataset to avoid modifying the original
data_transformed = data.copy()

# Select categorical columns
categorical_cols = data_transformed.select_dtypes(include=['object', 'bool']).columns

# Apply Label Encoding to all categorical columns
label_encoders = {}
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    data_transformed[col] = label_encoders[col].fit_transform(data_transformed[col])

# Display the transformed dataset
data_transformed.head()
```

```python
#Correlation matrix 


data_transformed = data.copy()

# Select categorical columns
categorical_cols = data.select_dtypes(include=['object', 'bool']).columns

# Apply Label Encoding to all categorical columns
label_encoders = {}
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    data_transformed[col] = label_encoders[col].fit_transform(data_transformed[col])

# Compute the correlation matrix for the transformed dataset
correlation_matrix_transformed = data_transformed.corr()

# Display the correlation matrix (first few rows for brevity)
correlation_matrix_transformed.head()
```

```python

# Transform categorical columns into numerical ones using Label Encoding
data_transformed = data.copy()
categorical_cols = data.select_dtypes(include=['object', 'bool']).columns

# Apply Label Encoding to all categorical columns
label_encoders = {}
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    data_transformed[col] = label_encoders[col].fit_transform(data_transformed[col])

# Compute the correlation matrix
correlation_matrix_transformed = data_transformed.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix_transformed, annot=True, fmt=".2f", cmap="coolwarm", 
            cbar=True, square=True, linewidths=0.5, linecolor='black', annot_kws={"size": 10})

# Add title and labels
plt.title('Correlation Matrix Heatmap (Transformed Data)', fontsize=16, fontweight='bold', color='darkblue')
plt.xticks(fontsize=10, fontweight='bold', rotation=45, ha='right')
plt.yticks(fontsize=10, fontweight='bold', rotation=0)
plt.show()
```
